# Top 3 Chillyroom(凉屋游戏)

```python
source = requests.get("https://www.taptap.com/developer/11051/apps") 
```

```python
print(source.content) 
```

```python
page_html = bs.BeautifulSoup(source.content, "html.parser") 
```

```python
page_html
```

```python
news_containers = page_html.find_all("div", class_= "taptap-app-list") 
```

```python
print(news_containers)
```

```python
titles = []
urls = []
gametypes = []
scores = []

for container in news_containers:
    for link in container.find_all("a", class_="item-caption-title flex-text-overflow taptap-link"): 
        print("title:")
        print(link.text.strip())
        titles.append(link.text.strip()) 
        print("url:")
        print(link.get("href"))
        urls.append(link.get("href")) 
        
    for gametype in container.find_all("span", class_="item-caption-label"):
        print("gametype:")
        print(gametype.a.string)
        gametypes.append(gametype.a.string)
        
    for score in container.find_all("span", class_="caption-label-rating pull-right"):
        print("score:")
        print(score.span.string)
        scores.append(score.span.string)
```

```python
print(len(titles))
print(len(urls))
print(len(gametypes))
print(len(scores))
```

```python
datatop3 = {
    "titles":titles,
    "urls":urls,
    "gametypes":gametypes,
    "scores":scores
}
```

```python
datatop3
```

```python
import pandas as pd 
Top3_Chillyroom = pd.DataFrame.from_dict(datatop3, orient ="index").transpose()
print(Top3_Chillyroom.info())
```

```python
Top3_Chillyroom
```

```python
Top3_Chillyroom.to_csv("Top3_Chillyroom.csv")
```
